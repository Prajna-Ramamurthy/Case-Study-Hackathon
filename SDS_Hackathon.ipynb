```python
import pandas as pd
df = pd.read_csv("SDS_Hackathon_Dataset_30_11_22.csv")
df = pd.DataFrame(df)
print(df)
```

```python
# 1. Explore the high-level characteristics of the dataset using standard functions in the Pandas Python library. 
# How many students have been surveyed for the collection of the dataset? Analyze the set of attributes available in the dataset. 
# List out the type of each attribute: is it categorical or numerical? Is it nominal, ordinal, discrete, or continuous?

df_rows = df.shape[0]

print(df_rows,"students have been surveyed for the collection of the dataset.\n")
print(df.describe())

print("\nThe attributes - Active, Rest, Hgt and Wgt are numerical and all the values are of integer type.")
print("The attributes Smoke, Gender and Exercise are categorical.")
print("The first column contains the index which is in ascending order, hence it is ordinal.")
print("All the other columns are discrete.")
```

```python
# 2. A summary statistic quantitatively summarizes the data in a particular feature using a single number.
# There are two popular types of summary statistics: measures of central tendency and measures of dispersion. 
# For each attribute in the dataset, figure out the most effective measure of central tendency and state its value. 
# Also, determine the standard deviation and range of values in each column. 
# Hint: Think about whether the mean, median, or mode is a more insightful measure of central tendency 
# for a feature depending on its type (categorical or numerical).

print(df.describe())

Active = df["Active"]
Rest = df["Rest"]
Smoke = df["Smoke"]
Gender = df["Gender"]
Exercise = df["Exercise"]
Hgt = df["Hgt"]
Wgt = df["Wgt"]

print("The mean of Active attribute is ",Active.mean()) 
print("The mean of Rest attribute is ",Rest.mean()) 
print("The mode of Smoke attribute is ",Smoke.mode()[0]) # mode is taken for categorical variable 
print("The mode of Gender attribute is ",Gender.mode()[0]) # mode is taken for categorical variable 
print("The mode of Exercise attribute is ",Exercise.mode()[0]) # mode is taken for categorical variable 
print("The mean of Hgt attribute is ",Hgt.mean())
print("The mean of Wgt attribute is ",Wgt.mean()) 
```

```python
# 3. Dealing with missing data is a crucial step in the process of data analysis. 
# It can have a significant impact on the conclusions drawn from the data. 
# Determine the number of missing records in each column of the dataset. 
# Figure out the most effective way to deal with missing data in each column.
# Transform each column and make sure none of the columns have any missing values.

import numpy as np
missing_values = ['',' ']
df = df.replace(missing_values,np.NaN)
df["Active"] = df["Active"].replace(np.NaN,Active.mean())
df["Rest"] = df["Rest"].replace(np.NaN,Rest.mean())
df["Smoke"] = df["Smoke"].replace(np.NaN,Smoke.mode()[0])
df["Gender"] = df["Gender"].replace(np.NaN,Gender.mode()[0])
df["Exercise"] = df["Exercise"].replace(np.NaN,Exercise.mode()[0])
df["Hgt"] = df["Hgt"].replace(np.NaN,Hgt.mean())
df["Wgt"] = df["Wgt"].replace(np.NaN,Wgt.mean())
```

```python
print(df.isnull().sum())
print(df.describe())
```

```python
# 4. Determine the number of classes of every categorical variable in the dataset. 
# Plot bar graphs for all the categorical attributes to determine the frequency of the classes in each attribute. 
# For example, plot a bar chart to determine the number of smokers and non-smokers. Do so for all categorical variables.
# What inferences about the survey can you make based on the visualizations?

import matplotlib.pyplot as plt
import numpy as np
import collections

#Vertical Bar Graph
smoke = df["Smoke"].values.tolist()
labels_smoke = ['Non-smoker', 'Smoker']

c = collections.Counter(smoke)
c = sorted(c.items())
freq = [i[1] for i in c]

f, ax = plt.subplots()

plt.bar(labels_smoke, freq)
plt.title("Smokers vs Non-smokers")
plt.xlabel("Smoke")
plt.ylabel("Frequency")
ax.set_xticks(range(0, 2))
ax.set_xticklabels(labels_smoke)
plt.show()

# There are less smokers
```

```python
# 6. Visualize the active and resting pulse rates of all the participants using box plots. 
# What is the interquartile range of the active pulse rates? How does this compare to the IQR of the resting pulse rates? 
# Now, visualize the active and resting pulse rates when grouped by gender. 
# What inferences can you make from the graph? 
# Repeat the same exercise for both the pulse rates when grouped by the amount of exercise and the smoking preference, separately. 
# Draw inferences from all the graphs obtained.

import matplotlib.pyplot as plt
from matplotlib.pyplot import boxplot, show 
import numpy as np

active = df["Active"].values.tolist()
rest = df["Rest"].values.tolist()
values = [active, rest]
plt.boxplot(values, labels = ["Active", "Rest"])
plt.title("Active vs Rest")
plt.ylabel("Values")
plt.show()

 
# First quartile (Q1)
a_Q1 = np.percentile(active, 25, method = 'midpoint')
r_Q1 = np.percentile(rest, 25, method = 'midpoint')
# Third quartile (Q3)
a_Q3 = np.percentile(active, 75, method = 'midpoint')
r_Q3 = np.percentile(rest, 75, method = 'midpoint')
  
# Interquaritle range (IQR)
a_IQR = a_Q3 - a_Q1
r_IQR = r_Q3 - r_Q1

print("Inter-quartile range for Active Pulse Rate :",a_IQR)
print("Inter-quartile range for Resting Pulse Rate :",r_IQR)
print("Inter-quartile range for Active Pulse Rate is more than that for Resting Pulse Rate")
```

```python
from scipy.stats import pearsonr
l = []
for i in list:
    for j in list:
        if(i!=j):
            list1 = df[i]
            list2 = df[j]
            corr, _ = pearsonr(list1,list2)
            print(i,"to",j,corr)
            if corr not in l:
                l.append(corr)
print(max(l))

print("By comparing the values, Height to Weight plot has the highest linear relationship.")
```

```python
# 8. Run a simple linear regression model for Active (the active pulse rate) against Rest (the resting pulse rate). 
# Make sure to create proper training and test sets and evaluate the goodness of the trained model. 
# What is the R^2 score of your model? State the fitted model equation using the slope and intercept values.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
%matplotlib inline
```

```python
print(df.shape) 
print(df['Active'].value_counts()) 
```

```python
new_df = df 
print(new_df.shape) 
print(new_df.isnull().sum()) 
```

```python
new_df = new_df[['Active','Rest']]
X = np.array(new_df[['Active']]) 
y = np.array(new_df[['Rest']])
print(X.shape) 
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=15)
regressor = LinearRegression() 
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
print('R2 score: %.2f' % r2_score(y_test,y_pred)) 
print('Mean squared Error :',mean_squared_error(y_test,y_pred)) 
```
